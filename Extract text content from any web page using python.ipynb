```python
import urllib3
```

```python
http = urllib3.PoolManager()
urllib3.disable_warnings()
url = 'https://sites.google.com/view/lekshmanmj' # Paste link here
response = http.request('GET', url)
```

```python
response.data
```

```python
s=str(response.data)
b=''
e=''
text_list=[]
```

```python
s[30000:]
```

```python
for i in range(0,len(s)):
    if s[i:i+1] == '>':
        b=i
        #print(b)
    elif s[i:i+1] == '<':
        e=i
        #print(e)
    if b!='' and e!='':
        if s[b+1:e] !='':
            #print(s[b+1:e])
            text_list.append(s[b+1:e])
        b=''
        e=''
```

```python
print(text_list)
```

```python
# Search the beginning of the content you require 
# and then pass the complete string to index()
text_list.index('LEKSHMAN M. J.  ')
```

```python
# similarly repeat the same for end of the required content
text_list.index(' Place : Chennai                                                                                                                                                                                                     ')
```
